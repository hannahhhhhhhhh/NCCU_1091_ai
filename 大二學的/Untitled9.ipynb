```python
import requests
from pyquery import PyQuery as pq
response = requests.get("https://rankedboost.com/pokemon-go/pokedex/") #使用了分頁內的網址，只列出了分頁內的商品
```

```python

doc = pq(response.text)    # 丟進pyQuery裡面
# print(doc(".rankedBoostPokedex .pokemontypeImgs").text())
for each in doc(".rankedBoostPokedex .pokemontypeImgs").items():
    nextpage=pq(each("a").attr("href"))
    print(doc(nextpage).text())
```
